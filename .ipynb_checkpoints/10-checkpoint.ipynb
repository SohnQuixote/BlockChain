# 이더리움 구성요소

## 키/주소
### 개인키 - 무작위 256비트 논스 , 공개키 - ECDSA를 활용하여 개인키로부터 유도

## 계정
### 계정에 연산이 일어나면 상태전이가 발생
### 두가지 종류가 존재
### 외부 소유 계정 - 연관코드 x, 개인키 제어, 인간 사용자
### 계약 계정 - 연관코드 있음(튜링 완전성으로 인해 어떠한 복잡도든 가능 채굴노드에 의해 EVM에서 실행) , 사용자X , 다른계약 호출 가능

## 트랜잭션/메시지
### 메시지 호출 트랜잭션 - 계약 계정 사이에서 메시지를 전달
### 계약 생성 트랜잭션 - 새로운 계약 계정을 생성 -> 연관 코드가 있는 계정이 생성
### 트랜잭션 필드 - 논스, 가스(수수료) ...

### 메시지 - 두 계정 간에 전달되는 데이터와 값 
### 메시지는 계약에 의해 생성되는 반면, 트랜잭션은 외부 소유 계정에 의해 생성됨

### 트랜잭션 보조 상태 - 트랜잭션 실행 도중 생성됨 (자폭 설정 , 로그 모음, 환불 잔고, 변경된 계정) 

## 상태 저장
### 전역 상태 - 주소와 상태간의 매핑
### 계정 상태 - 논스(트랜잭션 전송할 때마다 증가), 잔고, 저장 루트(계정의 저장 내용 인코딩 트리루트 노드), 코드 해시(스마트 계약 해시)


## 트랜잭션 영수증
### 트랜잭션이 실행된 이후의 상태를 저장


## 이더리움 가상 머신
### 바이트코드 명령어를 실행해 시스템 상태를 변경하는 단순한 스택 기반 실행 머신
### 튜링 완전 기계이지만 실행할 때 가스가 필요
### 사용 가능한 두 종류의 저장소 - 메모리(바이트 배열 RAM), 블록체인(키-값 하드디스크)

### 코드를 실행하기 위한 환경이 필요 (실행 환경 변수)
#### 머신 상태, 반복자 함수...

## 스마트 계약
### 자체지원 계약(라이브러리 함수나 리눅스의 시스템 콜 같은 느낌)
#### 로컬 노드에서 실행
#### 해싱, 공개키 복구, 타원곡선 연산...
