## 비트코인 네트워크
### 노드들이 트랜잭션과 블록을 주고받는 P2P 네트워크

## 노드의 종류
### 풀노드
#### 지갑,채굴, 전체 블록체인 저장, 네트워크 라우팅 기능을 모두 수행
### 단순 지불 검증 노드
#### 지갑, 네트워크 라우팅 기능만 수행

## 이러한 노드의 종류의 구성에 따라 다른 비트코인 네트워크를 구성할 수 있다.

## TCP, IP와 같이 비트코인 네트워크도 프로토콜 메시지를 통해 통신한다

### 27가지 유형의 프로토콜 메시지를 상황에 맞게 적절하게 사용한다

### 모든 블록을 동기화 하기 위해서 블록 우선방식이나 헤더 우선방식을 사용한다. (최근에는 헤더 우선방식만을 사용함)

## wire shark를 통해 비트코인의 프로토콜 메시지 교환을 볼 수 있다.



## 지갑도 여러 매체들을 통해서 사용이 가능하다

## BIP를 통해 지속적으로 발전한다.
