# 이더리움 심화

## 프로그래밍 언어
### 솔리디티 -> 바이트 코드 -> EVM (스택 언어) 
### 산술,논리, 암호화, 환경정보, 블록정보, 스택, 메모리흐름 제어 등등을 스택 명령어 수준으로 제공함
### 비트코인 스크립트도 동일한 스택언어이나 반복을 제공하지 않아 튜링완전하지 않은데
### 이는 무한루프를 막기 위함
### 이더리움은 가스라는 자원을 명령어가 수행할 때 소모되게 하여 무한루프를 막음


## 블록/블록체인
### 블록 -> 블록헤더, 트랜잭션 리스트, 옴머/엉클 헤더리스트
### 옴머/엉클 -> 채굴자가 채굴하려 했으나 실패한 무효블록 -> 보상 제공
### 블록헤더 -> 다양한 정보를 해싱함수를 통해 제공하며 올바른 정보를 저장하기 위한 목적을 지님
### 블록 유효성 검사 -> 이전블록 존재/유효 , 타임 스탬프 유효, 엉클/트랜잭션 정합성 만족
### 블록 종결 -> 블록 유효성 검사를 마친 뒤 채굴자에게 보상을 적용

### 지원 프로토콜 -> p2p 메시징 - 위스퍼 p2p저장 - 스웜
### PoW가 지속적으로 이어지면 난이도가 기하급수적으로 증가하여 채굴이 불가능한 시점이 올 예정
### 그 때 PoS로 채굴 메커니즘을 변경
