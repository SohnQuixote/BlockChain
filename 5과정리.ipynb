# 비트코인
## PoW (Proof of Work)
### 작업 증명 실제로 채굴할 때 해싱퍼즐을 하는 과정을 거침

## 머클트리, 해시함수, 공개 키 암호화, 디지털 서명

## 타인에게 돈 지불
### 1) 원하는 사람의 지불주소를 사용
### 2) 몇가지 규칙에 따라 트랜잭션을 구성해, 비트코인 네트워크에 브로드 캐스팅. 브로드캐스팅 하기전, 개인키를 사용해 디지털 서명
### 3)이후 채굴자가 검증후 블록체인에 블록을 추가함 채굴자는 수수룔르 지급받음 

## 트랜잭션 전주기
### 사용자가 트랜잭션 전송
### 보내는 사람이 개인키 사용해서 서명
### 채굴 노드가 이 트랜잭션을 검증하여 채굴할 다음 블록에 포함
### 채굴자에게 보상으로 새로운 코인 생성
### 채굴자가 해싱퍼즐의 해를 구하면 새로 채굴한 블록을 네트워크에 브로드캐스팅
### 각 노드들은 블록을 검증하고 승인 생성
### 마지막으로 받는 사람의 지갑에 승인, 승인 세번 후 트랜잭션 확정

## 스크립트 언어
### 비트코인에서는 간단한 스택 기반언어인 스크립트를 활용해 지출, 전송 과정을 트랜잭션내에 기술함
## 블록체인

## 해싱퍼즐
### 무차별 대입말고는 풀 방법이 없는 퍼즐을 활용 채굴의 난이도는 더욱더 올라갔음

## 트랜잭션의 구조

### 메타데이터
### 입력 - 로킹 해제 스크립트
### 출력 - 로킹 스크립트

###  많이 사용되는 스크립트 명령어
### OP_CHECKSIG, OP_EQUAL, OP_DUP, OP_HASH160, OP_VERIFY, OP_EQUALVERIFY, OP_CHECKMULTISIG

## 스크립트 사용 예시
### 지출하려면 로크를 해제해야 함
### 로크를 해제하려는 사용자가 로킹해제 스크립트를 함
### --------------------------보내는 사람----------------------------
### 출력의 일부로 로킹 스크립트가 들어가며 로킹 스크립트는 출력을 지출하려면 충족해야 하는 조건을 지정
### 로킹스크립트로 출력에 로크를 걸고 조건이 충족되면 출력의 로크가 해제


## 트랜잭션 유형
### P2PKH
### P2SH
### P2PK
### 입력, 출력 스크립트의 명령어에 따라서 다른 트랜잭션의 유형
### 마치 이더리움의 스마트 계약과 비슷한 개념이지만 간단한 스택 명령어의 수준밖에 못내림
